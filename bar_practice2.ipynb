```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ['PYTHON','C','C++','JAVA']
P_2022 = [45,10,15,30]
P_2023 = [50,5,10,35]

n = len(x)
width = 0.2
b1 = np.arange(n)
b2 = [x + width for x in b1 ]

plt.bar(b1,P_2022,width,color='red',edgecolor = 'black',label='2022')
plt.bar(b2,P_2023,width,color='g',edgecolor='black',label='2023')

plt.title('')

plt.xticks([s + width for s in range(n)],x)
plt.legend()
plt.show()
```
