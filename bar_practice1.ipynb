# Bar plot

```python
import matplotlib.pyplot as plt

x = ['python','c',"c++",'java']
y = [85,60,72,80]
plt.xlabel('languages',fontsize=15)
plt.title('VIVEK' ,fontsize=15)
plt.ylabel('popularity',fontsize=15)

c = ['red','blue','yellow','green']

plt.bar(x,y,width=0.2,color=c,edgecolor="black",linewidth=2,label=['python','c','c++','java'])
plt.legend()
plt.show()
```

```python
names = ["Python","C","C++","Dart","Java","JS"]
P_2021 = [90,40,50,65,70,70]
P_2022 = [92,20,30,60,75,80]
P_2023 = [95,30,35,68,75,82]

width = 0.2
n = len(names)

p1 = np.arange(n)
p2 = [names + width for names in p1]
p3 = [names + width for names in p2]

plt.title('Popularity Report',fontsize=15)
plt.xlabel('Languages',fontsize=15)
plt.ylabel('Popluarity count : ', fontsize=15)


plt.bar(p1,P_2021,width,color='r',edgecolor='black',label=2021)
plt.bar(p2,P_2022,width,color='y',edgecolor='black',label=2022)
plt.bar(p3,P_2023,width,color='g',edgecolor='black',label=2023)

plt.xticks([s + width for s in range(n)],names)
plt.legend()
plt.show()

```
