```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ['Python','C','C++','Java']
y1 = [9,3,5,7]
y2 = [10,2,4,7.5]

width=0.2
n = len(x)
blk = 'black'

p = np.arange(n)
p1 = [x + width for x in p]


plt.title("title")
plt.xlabel('X label')
plt.ylabel("Y label")


plt.bar(p,y1,width,edgecolor=blk,label='Y 1')
plt.bar(p1,y2,width,edgecolor=blk,label='Y 2')

plt.xticks([d + width for d in range(n)],x)
plt.legend()
plt.show()
```
